## Log plot

```python
import numpy as np


fig, ax = plt.subplots()

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
n, bins, patches = ax.hist(x, 50, density=1, facecolor='g', alpha=0.75)


ax.set_xlabel('Smarts')
ax.set_ylabel('Probability')
ax.set_title('Histogram of IQ')
ax.text(60, .025, r'$\mu=100,\ \sigma=15$')
ax.axis([40, 160, 0, 0.03])
ax.grid(True)

metadata = {"caption": "The Caption", "vars": ["a", "b"], "config": {"x": 1, "y": 2}}
vm.log_figure(fig, key="matplot", metadata=metadata, run_cuid=run_cuid)

```
